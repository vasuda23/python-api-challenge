```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Import API key
from api_keys import g_key
```

```python
cities_df = pd.read_csv('Output/weather_data.csv').reset_index(drop=True)
cities_df.head()

cities_df.columns

cleaned_cities = cities_df.drop(['Unnamed: 0'], axis=1)
cleaned_cities.head()
```

```python
gmaps.configure(api_key=g_key)
```

```python

locations = cleaned_cities[["Latitude", "Longitude"]].astype(float)
humidity_rate = cleaned_cities["Humidity (%)"].astype(float)

fig = gmaps.figure()

heat_layer = gmaps.heatmap_layer(locations, weights=humidity_rate, dissipating=False, max_intensity=250, point_radius=4)

fig.add_layer(heat_layer)

fig
```

```python
ideal_cities = cleaned_cities.loc[(cleaned_cities["Temperature (F)"] < 70) & (cleaned_cities["Temperature (F)"] > 50) & (cleaned_cities["Wind Speed (MPH)"] < 5) & (cleaned_cities["Cloudiness (%)"] > 50)]
ideal_cities_final = ideal_cities.reset_index()
ideal_cities_final.head()
```

```python

hotel_df = ideal_cities_final.copy()
hotel_df['Hotel Name'] = ''
```

```python
params = {
    "radius": 5000,
    "types": "hotel",
    "keyword": "hotel",
    "key": g_key
}

for index, row in hotel_df.iterrows():
    lat = row["Latitude"]
    lng = row["Longitude"]

    params["location"] = f"{lat},{lng}"

    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

    name_address = requests.get(base_url, params=params).json()
    
    try:
        hotel_df.loc[index, "Hotel Name"] = name_address["results"][0]["name"]
        print("fetching data for {lat},{lan}")
    except (KeyError, IndexError):
        print("No data....")
        
```

```python
hotel_df.head()
cleaned_hotel_df = hotel_df.reset_index(drop=True)
cleaned_hotel_df.drop(['index'], axis=1)
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in cleaned_hotel_df.iterrows()]
locations = hotel_df[["Latitude", "Longitude"]]
```

```python

# Add marker layer ontop of heat map
hotel_layer = gmaps.marker_layer(
    locations,
    info_box_content=hotel_info)

fig = gmaps.figure()

fig.add_layer(hotel_layer)
fig.add_layer(heat_layer)
    
# Display figure
fig
```
